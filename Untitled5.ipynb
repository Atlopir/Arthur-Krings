```python





import random


def flip_coin():
    coin = ''
    flip = random.randint(1, 10)
    if flip <= 5:
        coin = 'H'
    else:
        coin = 'T'
    return coin


def get_steps():
    steps = random.randint(1, 50)
    return steps

position = 50
left_wall = 0
right_wall = 100


for _ in range(5):
    # Flip a coin
    coin = flip_coin()


    steps = get_steps()


    print(coin, steps)


    if coin == 'H':
        position += steps
    else:

        position -= steps

    if position <= left_wall or position >= right_wall:
        print('You win!')
        break

else:
    print('You are at position', position, 'in the room.')

    position = 50

counter = 0

while position > left_wall and position < right_wall:

    steps = get_steps()


    coin1 = flip_coin()

    coin2 = flip_coin()

    print(steps, coin1, coin2)

    if coin1 == coin2:
        steps *= 2


    if coin1 == 'H':
        position += steps
    else:
        position -= steps

    #
    counter += 1

# The following statement does not need to be changed.
print('You reached the wall after', counter, 'chances.')

```
