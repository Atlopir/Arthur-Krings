```python
import random

# This function simulates flipping a coin. It returns the string H for heads
# and T for tails
def flip_coin():
    coin = ''
    flip = random.randint(1, 10)
    if flip <= 5:
      coin = 'H'
    else:
      coin = 'T'
    return coin

def get_steps():
    steps = random.randint(1, 50)
    return steps

position = 50
left_wall = 0
right_wall = 100

coin1 = flip_coin()
coin2 = flip_coin()


print('First toss:', coin1, 'Second toss:', coin2)

is_head1 = coin1 == 'H'
is_head2 = coin2 == 'H'
print(is_head1, is_head2)








steps_right = get_steps()
steps_left = get_steps()


if is_head1 and is_head2:

    position += steps_right
else:

    position -= steps_left

if position < left_wall:
    position = left_wall
elif position > right_wall:
    position = right_wall
elif is_head1 or steps_right > steps_left:

    position += (2 * steps_right) - steps_left
else:

    position += (2 * steps_left) - steps_right


if position < left_wall:
    position = left_wall
elif position > right_wall:
    position = right_wall

print("Your final position is:", position)



```
